```python
def court_init(newArray):
    ac=bc=cc=0 #counters for mini tournament
    a_final=b_final=c_final=0 #counters for final court allocation'
    bool_sum = np.sum(newArray.astype(bool), axis=0) #counts the number of true values in each column

    A_count=int(input("How many courts for session A?"))
    B_count=int(input("How many courts for session B?"))
    C_count=int(input("How many courts for session C?"))
    
    # Initialize the court arrays
    A_court = ['Coaching'] + ['Intermediate/Advanced'] * (A_count - 1)
    B_court = ['Coaching'] + ['Intermediate/Advanced'] * (B_count - 1)
    C_court = ['Intermediate/Advanced'] * (C_count)

    s1=s2=s3=0

    # Assign courts for mini tournament
    for i in newArray:
        if i[2] == True and i[3] == True:
            s1=s1+1
        if i[2] == True and i[4] == True:
            s2=s1+1
        if i[2] == True and i[5] == True:
            s3=s3+1
    
    if s1 >= 2: # if mini tournament and session a is true for 2 or more people
        A_court[1] = 'Mini Tournament'
        ac=1
    if s2 >= 2: #checks if more than 2 people wanna play the tournament
        B_court[1] = 'Mini Tournament'
        bc=1
    if s3 >= 2: 
        C_court[0] = 'Mini Tournament'
        cc=1

    a_final=(A_count-1)*4
    b_final=(B_count-1)*4
    c_final=(C_count)*4

    # print(ac," ",bc," ",cc)
    # print(a_final," ",b_final," ",c_final)
    
    # Assign remaining courts if there are leftover courts
    if A_count>1:
        if not (A_count==2 and A_court[1]=='Mini Tournament'):
            A_court[A_court.index('Intermediate/Advanced')] = 'Beginner/Intermediate'
    
    if B_count>1:
        if not (B_count==2 and B_court[1]=='Mini Tournament'):
            B_court[B_court.index('Intermediate/Advanced')] = 'Beginner/Intermediate'

    if C_count>1:
        if not (C_count==2 and C_court[0]=='Mini Tournament'):
            C_court[C_court.index('Intermediate/Advanced')] = 'Beginner/Intermediate'

    return A_court, B_court, C_court,ac,bc,cc,a_final,b_final,c_final

#We are assuming that there will not be a case where there is only 1 court available. 
```
